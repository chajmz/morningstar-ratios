```
list_df = []
cols_w = ['Category','2019','Average 5Y','Average 10Y','Symbol','Name Stock']


value_to_find = ['Profitability | Net Margin %','TC Net Income %','Net Earning TC Annuel >7%','Gross Margin %','Total Assets',
                'Margins % \of Sales | R&D',"Profitability | Return on Assets %","Profitability | Return on Equity %",
                "Current Ratio","Earnings per Share"]
                
rows_to_create = ['Net Earning to Revenue','Net Earning TC Annuel >7%','Net Earning TC Annuel 10y','Gross Margin>40%','Total Current Assets',
                'R&D % Of Gross Profit faible','Return On Assets > 7%','Return on Equity >25%',
                'Current Ratio','EPS']

value_to_find_new = ['Total Assets in Mil','Short/ Long Term Debt','Long Term Debt / Net Income','Total Liabilities/Total Equity']
rows_to_create_new = ['Total Assets','Short/ Long Term Debt','Long Term Debt / Net income < 4','Total Liabilities / Total Equity < 0,8']

def create_warren_simple(row,value_find,rows_create):
  value_2019 = 0
  average_5y = 0
  average_10y = 0
  symbol = row['symbol']
  name_stock = row['name_stock']
  for i in range (0,len(value_find)):  
      if row['year'] == value_find[i]:
          value_2019 = row['2019-12']
          average_5y = row["2015-12":"2019-12"].to_numpy()
          average_10y = row["2010-12":"2019-12"].to_numpy()
          a5 = average_5y[average_5y.nonzero()].mean()
          a10 = average_10y[average_10y.nonzero()].mean()

          list_df.append([rows_create[i],value_2019,a5,a10,symbol,name_stock])
```

```
## Helper function : calcule nouvelles valeurs pour tableau 
def compute_ratio_sga(df,var1, var2, name):
  row1 = df[df.year== var1]
  row2 = df[df.year== var2]
  list_df = [name]
  for i in range(1,11):
    ratio = row1.iloc[0][i] / row2.iloc[0][i]
    list_df.append(ratio)
  section = row1.iloc[0]['Section']
  symbol = row1.iloc[0]['symbol']
  name_stock = row1.iloc[0]['name_stock']
  list_df.append(np.nan)
  list_df.append(section)
  list_df.append(name_stock)
  list_df.append(symbol)
  return list_df
  

def compute_ratio_book_value(df,var1, var2,var3, name):
  row1 = df[df.year== var1]  #Total Stakeholders equity % total liabilities & equities
  row2 = df[df.year== var2]  #Book Value per Share
  row3 = df[df.year== var3]  #Shares Mil
  list_df = [name]
  for i in range(1,11):
    ratio = (row2.iloc[0][i]*row3.iloc[0][i] )/ (row1.iloc[0][i]/100)
    list_df.append(ratio)
  section = row1.iloc[0]['Section']
  symbol = row1.iloc[0]['symbol']
  name_stock = row1.iloc[0]['name_stock']
  list_df.append(np.nan)
  list_df.append(section)
  list_df.append(name_stock)
  list_df.append(symbol)
  return list_df

def compute_ratio_debt(df,var1, var2, name):
  row1 = df[df.year== var1]
  row2 = df[df.year== var2]
  list_df = [name]
  for i in range(1,11):
    ratio = row1.iloc[0][i] / row2.iloc[0][i]
    list_df.append(ratio)
  section = row1.iloc[0]['Section']
  symbol = row1.iloc[0]['symbol']
  name_stock = row1.iloc[0]['name_stock']
  list_df.append(np.nan)
  list_df.append(section)
  list_df.append(name_stock)
  list_df.append(symbol)
  return list_df


def compute_ratio_long_term_debt(df,var1, var2, name):
  row1 = df[df.year== var1]
  row2 = df[df.year== var2]
  list_df = [name]
  for i in range(1,11):
    ratio = row1.iloc[0][i] * row2.iloc[0][i] / 100    
    list_df.append(ratio)
  section = row1.iloc[0]['Section']
  symbol = row1.iloc[0]['symbol']
  name_stock = row1.iloc[0]['name_stock']
  list_df.append(np.nan)
  list_df.append(section)
  list_df.append(name_stock)
  list_df.append(symbol)
  return list_df


def compute_ratio_long_term_net_income(df,var1, var2, name):
  row1 = df[df.year== var1]
  row2 = df[df.year== var2]
  list_df = [name]
  for i in range(1,11):
    ratio = row1.iloc[0][i] / row2.iloc[0][i]  
    list_df.append(ratio)
  section = row1.iloc[0]['Section']
  symbol = row1.iloc[0]['symbol']
  name_stock = row1.iloc[0]['name_stock']
  list_df.append(np.nan)
  list_df.append(section)
  list_df.append(name_stock)
  list_df.append(symbol)
  return list_df


def compute_ratio_liability_equity(df,var1, name):
  row1 = df[df.year== var1]
  list_df = [name]
  for i in range(1,11):
    ratio = (row1.iloc[0][i]/100) / (1-(row1.iloc[0][i])/100)  
    list_df.append(ratio)
  section = row1.iloc[0]['Section']
  symbol = row1.iloc[0]['symbol']
  name_stock = row1.iloc[0]['name_stock']
  list_df.append(np.nan)
  list_df.append(section)
  list_df.append(name_stock)
  list_df.append(symbol)
  return list_df
```

```
global_rev_100 = []

def apply_base(row):
  new_base = []
  taux_croissance_net_income = []
  taux_croissance_10y = []
  str_year = row['year']
  if row['year'] == "Revenue":  ### Revenue base 100

    new_base.append("Revenue B100%")
    base_100 = row[1]
    new_base.append(100)
    for current_revenue in row[2:12]:
      new_base.append(round(float(current_revenue)/base_100*100,1))

    
    new_base.append(row['Section'])
    new_base.append(row['name_stock'])
    new_base.append(row['symbol'])
    global_rev_100.append(new_base)

  elif row["year"] == "Net Income":   #pareil qu'au dessus on change juste le nom de la ligne en net income base 100

    new_base.append("Net Income Base 100%")
    taux_croissance_net_income.append("TC Net Income %")
    taux_croissance_10y.append("TC Net Income 10y ")
    taux_croissance_10y.append(np.nan)
    taux_croissance_net_income.append(np.nan)
    base_100 = row[1]
    new_base.append(100)
    previous_taux_croissance = base_100
    net_income_y0 = base_100
    index_y = 1

    for current_value in row[2:12]:
      index_y += 1
      new_revenue = float(current_value)/base_100*100 #calcul revenue base 100
      new_base.append(new_revenue) 
      new_taux = ((float(current_value) - previous_taux_croissance)/previous_taux_croissance)*100 # calcul variation taux net income
      taux_croissance_net_income.append(new_taux)
      previous_taux_croissance = current_value
      tc_net_income_10y = 100*(pow(float(current_value)/net_income_y0,1/(index_y-1))-1)
      taux_croissance_10y.append(tc_net_income_10y)

    taux_croissance_net_income.append(row['Section'])
    taux_croissance_net_income.append(row['name_stock'])
    taux_croissance_net_income.append(row['symbol'])
    
    
    
    taux_croissance_10y.append(row['Section'])
    taux_croissance_10y.append(row['name_stock'])
    taux_croissance_10y.append(row['symbol'])

    global_rev_100.append(taux_croissance_net_income)
    global_rev_100.append(taux_croissance_10y)
    

  elif row['year'] == "Earnings per Share":

    new_base.append("Earnings per Share B100%")
    base_100 = row[1]
    new_base.append(100)
    for current_revenue in row[2:12]:
      new_base.append(round(float(current_revenue)/base_100*100,1))
    new_base.append(row['Section'])
    new_base.append(row['name_stock'])
    new_base.append(row['symbol'])
    
    
    global_rev_100.append(new_base)

  elif row['year'] == "Dividends":

    new_base.append("Dividends B100%")
    base_100 = row[1]
    if np.isnan(base_100):
      new_base.append(base_100)
    else:
      new_base.append(100)
    for current_revenue in row[2:12]:
      div_value = float(current_revenue)
      if np.isnan(base_100):
        base_100 = div_value
      new_base.append(round(div_value/base_100*100,1))
    
    new_base.append(row['Section'])
    new_base.append(row['name_stock'])
    new_base.append(row['symbol'])
    global_rev_100.append(new_base)

  elif row['year'] == "Earnings per Share":

    new_base.append("Earnings per Share B100%")
    base_100 = row[1]
    new_base.append(100)
    for current_revenue in row[2:12]:
      new_base.append(round(float(current_revenue)/base_100*100,1))
    new_base.append(row['Section'])
    new_base.append(row['name_stock'])
    new_base.append(row['symbol'])
    
    
    global_rev_100.append(new_base)      
     


#test sur apple
df_aapl = df[df['symbol'] == 'AAPL'] 
```
